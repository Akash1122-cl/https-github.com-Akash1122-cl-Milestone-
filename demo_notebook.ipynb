```python
# Test the live demo
import requests
import json

# Test basic factual query
response = requests.post(
    'http://127.0.0.1:8000/api/chat/query',
    json={
        'thread_id': 'test-session',
        'query': 'What is the expense ratio for Nippon India Large Cap Fund?',
        'scheme_name': None
    }
)

if response.status_code == 200:
    data = response.json()
    print(f'Answer: {data["answer"]}')
    print(f'Citation: {data["citation"]}')
    print(f'Is Advisory: {data["is_advisory"]}')
else:
    print('Error:', response.status_code)

# Test metrics API
metrics_response = requests.get('http://127.0.0.1:8000/api/metrics/funds')
if metrics_response.status_code == 200:
    funds = metrics_response.json()
    print(f'Total funds: {len(funds)}')
    for fund in funds:
        print(f'{fund["fund_name"]}: {fund["expense_ratio"]}')
```
